```python
import dowker_trajectories
from dowker_trajectories.Data import load_lorenz63, load_lorenz96, load_cdv, load_jetlat
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import gudhi as gd
```

```python
l63 = load_lorenz63()
l96 = load_lorenz96()
cdv = load_cdv()
jetlat = load_jetlat()
datasets = [l63, l96, cdv, jetlat]
for d in datasets:
    print(d.shape)
```

```python
from dowker_trajectories import Trajectory
traj63 = Trajectory(l63[:,0:100])
bins63 = traj63.set_bins(15)
binseq63 = traj63.bin_sequence()
A63, verts63, counts63 = traj63.adjacency()
```

```python
binloc63 = traj63.bin_centers()
```

```python
from dowker_trajectories.Digraph import Digraph
G63 = Digraph(A63, verts63, counts63, loops = False)
```

```python
from dowker_trajectories.viz import graph_graph
fig = graph_graph(G63, traj63.bin_centers(), arrows=True)
```

```python
fig.show()
```

```python
D63 = G63.distance_matrix()
```

```python
from dowker_trajectories import DowkerComplex
MNeighbor = DowkerComplex(D63).create_simplex_tree(filtration='Sublevel', max_dimension=2)
dgms = MNeighbor.persistence() #this is the gudhi built in persistence function
```

```python
gd.plot_persistence_diagram(dgms)
plt.show()
```

```python
gd.plot_persistence_barcode(dgms)
plt.show()
```
